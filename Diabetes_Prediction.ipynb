```python
X = diabetes_dataset.drop(columns ='Outcome',axis = 1)
Y = diabetes_dataset['Outcome']
```

###Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
standardized_data
#what this standardization has done is that
#it has trasformed our data from various inconsistent values to
#some values between a specific range it might be using some mathematical
#formula to do so but we don't need to get into that.
#This is done in order for our machine learning model to predict the values properly
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

###Train Test Split





```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
#In the above line Stratify means maintaining the proportion of Y in Y_train and Y_test
#Such that the outcome values are not disproportionate, which will help in better prediction
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape,
```

###Training the model

####Support Vector Machine



It is a Supervised Learning Classification model which uses a Hyperplane which is a simple line to separate the class labels. Once separated it tries to maximize the distance between the labels and during the prediction it puts the data o/p into different planes accordingly. For linearly inseperable(meaning they are placed in such a way that they cannot be separated in a two dimensional space with the help of a line) datapoints it uses kernel parameter which takes the datapoints into a higher dimension, and then separates the points








```python
#loading the model function into a variable
#Support Vector classifier
classifier = svm.SVC(kernel = 'linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

### Model Evaluation

####Accuracy Score

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(f"The accuracy of our prediction model on the training data is {training_data_accuracy.round(2)} %")
```

Now let's understand what we have done in the above step:-

Firstly, we used the predict function on the classifier variable which contains the svm model, and in that predict function we inputted the X_train data in order to make the o/p prediction of it and stored it in the X_train_prediction variable.

Moreover, we predicted the accuracy score of our prediction by comparing the lables or o/p stored in the X_train_prediction with the original Y_train i.e., the o/p.

#### Now Predicting the test data and finding its accuracy

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(f"The accuracy of our prediction model on the test data is {test_data_accuracy.round(2)} %")
```

###Building a Predictive system

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

#changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardizing the input_data
std_data = scaler.transform(input_data_reshaped)
print(std_data)
```

```python
#Now predicting the data on the basis of input_data that is standardized
prediction = classifier.predict(std_data)
if(prediction == 0):
  print("The person is non-diabetic.")
else:
  print("The person is diabetic.")
```
